## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
data.columns
```

```python
desired = pd.read_sql_query('select * from DatabaseDesireNextYear',conn)
desired.head()
desired.drop_duplicates(inplace=True)
# desired.sort_values('Respondent',ascending=False).head()
```

```python
plt.figure(figsize=(10,10))
desired_gr = desired.groupby('DatabaseDesireNextYear')['Respondent'].count()
required = desired_gr.sort_values(ascending=False).to_frame().reset_index().head()
required['Respondent'].plot(kind='pie',autopct ='%1.2f%%',labels = required.loc[:5,'DatabaseDesireNextYear'])
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
desired = pd.read_sql_query('select * from LanguageDesireNextYear',conn)
desired.head()
desired.drop_duplicates(inplace=True)

# desired.sort_values('Respondent',ascending=False).head()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

aged = data[(data['Age']>=30) & (data['Age']<=35)]
one = aged.groupby('Age')['WorkWeekHrs'].median().to_frame().reset_index()
two = aged.groupby('Age')['CodeRevHrs'].median().to_frame().reset_index()
one.head()
plt.bar(x= one.Age,height=one.WorkWeekHrs,label = 'WorkWeekHrs')
plt.bar(x= two.Age,height=two.CodeRevHrs,bottom=one.WorkWeekHrs,label = 'CodeRevHrs')
plt.title('median WorkWeekHrs and CodeRevHrs for age groups')
plt.legend(loc = 'upper right',bbox_to_anchor=(1.4,1))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data for the age group 30 to 35
aged = data[(data['Age'] >= 30) & (data['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs for each age
median_workweekhrs = aged.groupby('Age')['WorkWeekHrs'].median().to_frame().reset_index()
median_coderevhrs = aged.groupby('Age')['CodeRevHrs'].median().to_frame().reset_index()

# Merge the two dataframes on 'Age'
merged_data = pd.merge(median_workweekhrs, median_coderevhrs, on='Age', suffixes=('_WorkWeek', '_CodeRev'))

# Plotting the stacked bar chart
plt.figure(figsize=(10, 7))
plt.bar(merged_data['Age'], merged_data['WorkWeekHrs'], label='WorkWeekHrs')
plt.bar(merged_data['Age'], merged_data['CodeRevHrs'], bottom=merged_data['WorkWeekHrs'], label='CodeRevHrs')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()

# Display the chart
plt.show()
```
